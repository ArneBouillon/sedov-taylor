```python
import yt
print(yt.__version__) # 3.6.0 or higher
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
```

## 1D

```python
datfile = 'output/sedov_taylor_0080.dat'
ds = yt.load(datfile)
```

```python
all_data_level_0 = ds.covering_grid(level=0, left_edge=[0, 0, 0], dims=ds.domain_dimensions)
x = np.linspace(0, 1, ds.domain_dimensions[0])
plt.plot(x, np.log10(np.squeeze(all_data_level_0['density']) / ds.unit_registry.lut['code_density'][0]), '-')
plt.plot(x, np.log10(np.squeeze(all_data_level_0['thermal_pressure']) / ds.unit_registry.lut['code_pressure'][0]), '--')
plt.plot(x, np.squeeze(all_data_level_0['velocity_magnitude']) / ds.unit_registry.lut['code_velocity'][0], '-.')
plt.ylim(-3, 1)
plt.xlabel('r')
plt.ylabel('Value')
plt.legend(['Density (log)', 'Pressure (log)', 'Velocity (linear)'], bbox_to_anchor=(0.8, 1.6))
plt.savefig('plots-8.png', bbox_inches='tight')
```
